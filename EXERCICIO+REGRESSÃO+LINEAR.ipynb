```python
import pandas as pd # conventional alias 
from sklearn.datasets import load_boston
```

```python
dataset = load_boston()
```

```python
print(dataset.data.shape)
```

```python
print(dataset.keys())
```

```python
print(dataset.feature_names)
```

```python
print(dataset.DESCR)
```

```python
bos = pd.DataFrame(dataset.data)
print(bos.head())
```

```python
bos.columns = dataset.feature_names
bos.head()
```

```python
print(dataset.target.shape)
```

```python
bos['PRICE'] = dataset.target
```

```python
print(bos.head())
```

```python
bos.describe()
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.pairplot(bos, x_vars=['CRIM', 'ZN','INDUS','CHAS', 'NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'
], y_vars='PRICE', size=7,aspect=0.7,kind='reg')
```

```python
features_cols = bos[['CRIM', 'ZN','INDUS','CHAS', 'NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'
]]
```

```python
X = features_cols
```

```python
X.head()
```

```python
print(type(X))
print(X.shape)
```

```python
y = bos['PRICE']
```

```python
y = bos.PRICE
```

```python
y.head()
```

```python
print(type(y))
print(y.shape)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
```

```python
print(linreg.coef_)
```

```python
list(zip(features_cols, linreg.coef_))
```

```python
y_pred = linreg.predict(X_test)
```

### - ROOT MEAN SQUARED ERROR

```python
from sklearn import metrics
```

```python
import numpy as np
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
features_cols = ['CRIM', 'ZN','INDUS','CHAS','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']

x = bos[features_cols]

y = bos.PRICE

X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=1)

linreg.fit(X_train, y_train)

y_pred = linreg.predict(X_test)

print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
